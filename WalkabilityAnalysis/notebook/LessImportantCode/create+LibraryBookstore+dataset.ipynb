```python
GPS = {}  # nested list
Location = {}
for name in data["name"]:
    location_boundry(name) 
```

```python
data['GPS_lng']=np.nan
data['GPS_lat']=np.nan

# save GPS
count = 0
for j in data['name']:
    for k,v in GPS.items():
        if j == k:
            data['GPS_lng'][count] = v[0]
            data['GPS_lat'][count] = v[1]
    count += 1
```
